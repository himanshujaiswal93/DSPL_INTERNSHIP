```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
# droping id column 
df.drop(columns=["Id"], inplace=True)
```

```python
df.info()
```

```python
print(df['Species'].value_counts())
```

```python
# visual EDA
# pair plot
sns.pairplot(df, hue="Species")
plt.show()
```

```python
# box plot for cheaking outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient="h")
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
# Species countplot
sns.countplot(x='Species', data=df)
plt.title("Species Distribution")
plt.show()
```

# Insights
No missing values.

Dataset is balanced across 3 species.

Features like petal length & width are strong indicators of species.

Few mild outliers in sepal width, but overall clean data.
